## Konstruktor und Destruktor

- Ein *Konstruktor* ist eine spezielle Methode (`__init__`), die beim Erzeugen eines Objekts aufgerufen wird
- Üblicherweise initialisiert der Konstruktor die (wichtigsten) Attribute eines Objekts
- Der *Destruktor* (`__del__`) wird aufgerufen, wenn die letzte Referenz auf ein Objekt "gelöscht" wird

```python
class A():
    a_counter = 0
    
    def __init__(self, value):
        self.value = value
        A.a_counter += 1

    def __del__(self):
        A.a_counter -= 1

a0 = A(0)
a1 = A(1)
a2 = A(2) 
a3 = A(3)
print(f"Es gibt {a0.a_counter} A's")
a2=42
print(f"Es gibt nun {a0.a_counter} A's")
del a0, a1, a2, a3
```

```python
class A():
    ''' Meine Klasse A'''
    the_as = []
    
    def __init__(self, value):
        self.value = value
        A.the_as.append(self)

    def __del__(self):
        A.the_as.remove(self)
        
a0 = A(0)
a1 = A(1)
a2 = A(2)
a3 = A(3)
a2 = 42

for a in A.the_as:
    print (a.value)
    
del a0, a1, a2, a3 
```

## Methoden

- ...sind Funktionen die *etwas mit Objekten einer Klasse tun*
- Jede Methode verwendet mindestens ein Objekt der Klasse
- Eine Methode ist im *Namensraum* der Klasse definiert
- Eine Methode kann mit `Klassenname.Methodenname(Objektreferenz)` aufgerufen werden

- Besser: Aufruf mit `Objektreferenz.Methodenname`


```python
class A():  
    def __init__(self, val):
        self.val = val
    def inc_and_print(self):
        self.val += 1
        print(f"Objekt {id(self)} mit Wert {self.val}")

a = A(2)
A.inc_and_print(a)
a.inc_and_print()
```

- Der erste Parameter eine Methode bezeichnet immer **das Objekt, auf das die Methode angewendet wird**
- Diesen Parameter nennt man auch *Selbstreferenz*
- Per Konvention wird diese Selbstreferenz `self` genannt
- Beim Aufruf einer Methode über die Punkt-Notation, wird das Objekt ("vor dem Punkt") als Argument für `self` eingesetzt

- Es gibt auch Funktionen, die zwar zu einer Klasse gehören, aber *nichts* mit Objekten der Klasse *tun*
- Solche Funktionen heißen *Klassenfunktion* oder *statische Methode*
- Um solche Funktionen auch über Objekte aufrufen zu können, muss man sie mit dem *Label* `@staticmethod` versehen

```python
class A():  
    def __init__(self, val):
        self.val = val

    @staticmethod
    def value_print(x):
        print(f"Ein Wert {x}")

a = A(2)
a.value_print(3)
```

## Vererbung

- Zentrales Konzept in der Objektorientierten Programmierung
- (Neue) Klassen können von bestehenden Klassen **abgeleitet** werden und **erben** dadurch alle Methoden und Attribute ihrer **Elternklasse** (auch *Basisklasse* genannt)

```python
class A():
    def __init__(self):
        self.x = 100
    def drucke_x(self):
        print("Mein x ist", self.x)

class B(A):
    pass

b = B()
b.drucke_x()
```

- Die abgeleitete Klasse kann für ihren eigenen Zweck angepasst und erweitert werden
- Es können neue Attribute und Methoden eingeführt werden, oder auch bestehende Attribute und Methoden von der Basisklasse überschrieben werden
- Ist eine aufgerufende Methode in der Klasse eines Objekts nicht vorhanden, wird nach der Methode in den Basisklassen gesucht
- Die erste *passende* Methode wird aufgerufen ($\rightarrow$ *Method Resolution Order*, MRO)
- Es kann mehrere Basisklassen geben ($\rightarrow$ *Mehrfachvererbung*)
- Um explizit auf Attribute und Methode der Basisklasse zuzugreifen, kann man die Funktion `super()` verwenden
- Mit `super().__init__()` kann der übergeordnete Konstruktor aufgerufen werden
- Sinnvoll, um doppelten Code zu vermeiden

```python
class A():
    classattrib = 'classattrib in der Klasse A'
    def __init__(self):
        self.x = 100
    def drucke_x(self):
        print("Mein x ist", self.x)

class B(A):
    classattrib = 'classattrib in der Klasse B'
    def __init__(self):
        super().__init__()
        print(super().classattrib)
        print(self.classattrib)

b = B()
b.drucke_x()
```

## Magic Methods und Magic Attributes

- Beispiel: Wenn ein Objekt erzeugt werden soll, wird immer die Methode `__new__` und dann `__init__` aufgerufen
- Alle *Magic Methods* und *Magic Attributes* sind in doppelten Unterstrichen `__`eingefasst
- Es gibt weitere *Magic Methods* mit denen Sie Operatoren *überladen* können ($\rightarrow$ Praktikumsaufgabe von heute)

```python
class Punkt():
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def set_label(self, label):
        self.label = label

class Punkt3D(Punkt):
    def __init__(self, x, y, z):
        super().__init__(x,y)
        self.z = z
    def __str__(self):
        res = ""
        if self.label:
            res += self.label + ": "
        res += f"({self.x}, {self.y}, {self.z})"
        return res

p1 = Punkt3D(2.0, -4.1, 3.3)
p1.set_label("Erster Punkt")
print(p1)
```

## Zugriffsregeln

- Wenn man ein Objekt (einer Klasse) verwendet, muss man i.d.R. nicht *alles über die Klasse wissen*
- Bestimmte Methoden und Attribute werden nur *innerhalb* der Klasse verwendet, aber nicht *von Außen*
- Viele objektorientierte Sprachen verwenden **Zugriffsmodifikatoren** (z.B. *public* und *private*) um die **Sichtbarkeit** von Methoden und Attributen einzugrenzen

- In Python gibt es solche Zugriffsmodifikatoren *nicht direkt*
- Über Namenskonventionen kann man ausdrücken, ob ein Attribut geschützt ist

- **Einfacher Unterstrich (`_x`):** Das Attribut `_x` *sollte* extern nicht direkt verwendet werden
- **Doppelter Unterstrich (`__x`):** Das Attribut `__x` *kann* extern nicht (ohne Weiteres) direkt verwendet werden

```python
class A():
    def __init__(self):
        self.x = 100
        self._x = 200
        self.__x = 300
    def __str__(self):
        return f"x = {self.x}, _x = {self._x}, __x = {self.__x}"

a = A()
print(a)
print(a.x)
print(a._x)
#a.__dict__
#print(a._A__x)
```

## Properties (<s>Getter und Setter</s>)

```python
class A:
    def __init__(self):
        self.__x = 100

    def getX(self):
        return self.__x

    def setX(self, val):
        if val < 0: return;
        self.__x = val
        
    x = property(getX, setX)
```

```python
a = A()
print(a.x)
a.x = 42
print(a.x)
a.x = -1
print(a.x)
a._A__x
```

```python
class A:
    def __init__(self):
        self._x = 100

    @property
    def x(self):
        return self._x

    @x.setter
    def x(self, wert):
        if wert < 0:
            return
        self._x = wert

a = A()
print(a.x)
a.x = 300
print(a.x)
a.x = -20
print(a.x)
```
